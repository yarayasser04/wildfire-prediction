```python
from google.colab import drive
import geopandas as gpd
import pandas as pd
import glob
from shapely.geometry import Point

drive.mount('/content/drive')

#Wildfire Dataset
wildfire_dataset = "/content/drive/MyDrive/mtbs_fod_pts_data/mtbs_FODpoints_DD.shp"
gdf_wildfire = gpd.read_file(wildfire_dataset)

#Weather Dataset
weather_dataset = "/content/drive/MyDrive/WeatherDataFromAcrossAlaska"
weather_files = glob.glob(weather_dataset + "/*.csv")
print(f"Found {len(weather_files)} weather files.")

#Combine all csv files from the weather dataset into one dataframe
df_list = []

for file in weather_files:
  df_weather = pd.read_csv(file)
  df_list.append(df_weather)

df_weather = pd.concat(df_list)
print(f"CSV (weather dataset) files combine shape: {df_weather.shape}")
df_weather.head()
```
